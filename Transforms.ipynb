```python
import os
from PIL import Image, UnidentifiedImageError
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms

# Define transformations
transform = transforms.Compose([
    transforms.RandomRotation(20),     
    transforms.RandomResizedCrop(256, scale=(0.8, 1.0)), 
    transforms.RandomHorizontalFlip(), 
    transforms.RandomVerticalFlip(),   
    transforms.ToTensor(),              
])

input_dir = 'C:/Users/HP/Documents/projects/NCS/Group_4_NCS/chest_xray/chest_xray/train/NORMAL'
output_dir ='C:/Users/HP/Documents/projects/NCS/Group_4_NCS/chestray/chest_xray/Augmented_data/NORMAL'


os.makedirs(output_dir, exist_ok=True)


allowed_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff']


for img_name in os.listdir(input_dir):
    img_path = os.path.join(input_dir, img_name)

   
    if not any(img_name.lower().endswith(ext) for ext in allowed_extensions):
        continue

    try:
        image = Image.open(img_path).convert('L')  # Convert to grayscale
    except UnidentifiedImageError:
        print(f"Cannot identify image file {img_path}")
        continue

    for i in range(5): 
        augmented_image = transform(image)

       
        augmented_image_pil = transforms.ToPILImage()(augmented_image)

       
        base_name, ext = os.path.splitext(img_name)
        augmented_image_name = f"{base_name}_aug_{i}{ext}"
        augmented_image_path = os.path.join(output_dir, augmented_image_name)
        augmented_image_pil.save(augmented_image_path)

        print(f"Saved {augmented_image_path}")
```
